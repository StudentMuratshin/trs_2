```python
import numpy as np
import matplotlib.pyplot as plt

def func(t, x):
    return (1-np.exp(-t))*np.sin(2*np.pi*x)+x**2

# Создаем массивы t и x для значений от -5 до 5 с шагом 0.1
t = np.arange(0, 5, 0.1)
x = np.arange(0, 5, 0.1)

# Вычисляем значения функции для каждой точки на сетке
Z = func(t, x)

# Строим 2D график функции
plt.plot(x, Z)
plt.xlabel('x')
plt.ylabel('u')
plt.show()
```
