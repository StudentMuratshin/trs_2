```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создаем данные для построения графика
t = np.linspace(0, 1, 100)
x = np.linspace(0, 1, 100)
t, x = np.meshgrid(t, x)
y = (1 - np.exp(-t)) * np.sin(2 * np.pi * x) + x ** 2

# Строим график
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(t, x, y)

# Настройка осей и заголовка
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_zlabel('y')
plt.title('(1-np.exp(-t))*np.sin(2*np.pi*x)+x**2')

# Отображаем график
plt.show()
```
