```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#    
data = np.genfromtxt('u1_out.csv', delimiter=' ')

#    
timesteps = data[:,0]
spatial_steps = data[:,1]
values = data[:,2]

#  3D 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(timesteps, spatial_steps, values)

#    
ax.set_xlabel(' ')
ax.set_ylabel('  ')
ax.set_zlabel('')

plt.show()
```
