```python
import numpy as np

#    
data = np.genfromtxt('u1_out.csv', delimiter=' ')

#     
time_steps = len(np.unique(data[:, 0]))
space_steps = len(np.unique(data[:, 1]))
solution = data[:, 2].reshape(time_steps, space_steps)

#    
x = np.linspace(0, 1, space_steps)
y = np.linspace(0, 1, time_steps)
X, Y = np.meshgrid(x, y)

#  3D 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(Y, X, solution)
ax.set_xlabel('Space')
ax.set_ylabel('Time')
ax.set_zlabel('Solution')
plt.show()
```

```python
import numpy as np

#    
data = np.genfromtxt('u1_out.csv', delimiter=' ')

#     
time_steps = len(np.unique(data[:, 0]))
space_steps = len(np.unique(data[:, 1]))
solution = data[:, 2].reshape(time_steps, space_steps)

#    
x = np.linspace(0, 1, space_steps)
y = np.linspace(0, 1, time_steps)
X, Y = np.meshgrid(x, y)

#  3D 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(X.flatten(), Y.flatten(), solution.flatten(), cmap='viridis', edgecolor='none')
ax.scatter(X.flatten(), Y.flatten(), solution.flatten(), c='r', marker='o')
ax.set_xlabel('Space')
ax.set_ylabel('Time')
ax.set_zlabel('Solution')
plt.show()
```
