```python
import numpy as np

# чтение данных из файла
data = np.genfromtxt('u1_out.csv', delimiter=' ')

# преобразование данных в трехмерный массив
time_steps = len(np.unique(data[:, 0]))
space_steps = len(np.unique(data[:, 1]))
solution = data[:, 2].reshape(time_steps, space_steps)

# создание сетки для графика
x = np.linspace(0, 1, space_steps)
y = np.linspace(0, 1, time_steps)
X, Y = np.meshgrid(x, y)

# построение 3D графика
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(Y, X, solution)
ax.set_xlabel('Space')
ax.set_ylabel('Time')
ax.set_zlabel('Solution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# читаем данные из файла
data = np.genfromtxt('u3_out.csv', delimiter=' ')

# разделяем данные на столбцы
timesteps = data[:,0]
spatial_steps = data[:,1]
values = data[:,2]

# создаем 3D график
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(timesteps, spatial_steps, values)

# добавляем метки к осям
ax.set_xlabel('Временные шаги')
ax.set_ylabel('Шаги по пространству')
ax.set_zlabel('Значения')

plt.show()
```

```python
import numpy as np

# чтение данных из файла
data = np.genfromtxt('u1_out.csv', delimiter=' ')

# преобразование данных в трехмерный массив
time_steps = len(np.unique(data[:, 0]))
space_steps = len(np.unique(data[:, 1]))
solution = data[:, 2].reshape(time_steps, space_steps)

# создание сетки для графика
x = np.linspace(0, 1, space_steps)
y = np.linspace(0, 1, time_steps)
X, Y = np.meshgrid(x, y)

# построение 3D графика
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(X.flatten(), Y.flatten(), solution.flatten(), cmap='viridis', edgecolor='none')
ax.scatter(X.flatten(), Y.flatten(), solution.flatten(), c='r', marker='o')
ax.set_xlabel('Space')
ax.set_ylabel('Time')
ax.set_zlabel('Solution')
plt.show()
```
